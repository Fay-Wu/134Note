# Updating your forked repo

- As we discussed in Lab 1 (Introduction to Git),  we 'forked' the course repo to our own Github account in order to have access to the course material.

- The first time that we forked the repo in question, the most updated version of the course repo was copied to your account.

- However, as the class progresses (and more material is added to the main repo), we want to be able to reflect those changes to our forked account. 

- To do so, we will follow the steps below:

## A. Configuring a remote for a fork

Before you can sync your fork with an upstream repository, you must configure a remote that points to the main repository in Git.

You must configure a remote that points to the `UCSB-PSTAT-134-234/Spring-2018` repository (the original course repo) in Git to sync changes you make in a fork with the original repository. 

To sync changes made in the original repository with the fork:

1. Open Terminal.

2. List the current configured remote repository for your fork.

    `git remote -v`

3. Specify a new remote upstream repository that will be synced with the fork.

    `git remote add upstream https://github.com/UCSB-PSTAT-134-234/Spring-2018.git`

4. Verify the new upstream repository you've specified for your fork.

    `git remote -v`

Source: https://help.github.com/articles/configuring-a-remote-for-a-fork/

## B. Syncing a fork

To sync a fork of a repository to keep it up-to-date with the upstream repository follow the next steps:

1. Open Terminal.

2. Change the current working directory to your local project.

3. Fetch the branches and their respective commits from the upstream repository. 

    `git fetch upstream`
    
    Commits to master will be stored in a local branch, `upstream/master`.
    
4. Check out your fork's local master branch.

    `git checkout master`
        
5. Merge the changes from `upstream/master` into your local master branch. This brings your fork's master branch into sync with the upstream repository, without losing your local changes.

    `git merge upstream/master`
    

## C. Push changes to your forked remote

Use `git push` to push commits made on your local branch to a remote repository.

`git push  remote master`
    
- You will be asked for your Github user name and password.


## D. Verify the changes in your forked repo

Go to Github, reload the website and check whether the version in your repo match the version on the main course repo.
